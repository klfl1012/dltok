```python
import matplotlib.pyplot as plt
from netCDF4 import Dataset

# --- Datei öffnen ---
file = "Data_files/BOUT.dmp.0.nc"
nc = Dataset(file)

# --- Variable 'n' auslesen und auf sinnvolle Form bringen ---
n = nc.variables['n'][:]      # (93, 12, 1, 1024)
n = n.squeeze()               # -> (93, 12, 1024)
print("Shape:", n.shape)

# --- Einen Zeitschritt auswählen, z.B. den letzten ---
frame = -1
data = n[frame, :, :]

# --- 2D-Plot ---
fig, ax = plt.subplots(figsize=(8, 4))
c = ax.imshow(
    data.T,
    origin="lower",
    aspect="auto",
    cmap="plasma"
)
fig.colorbar(c, ax=ax, label="normalized density n")
ax.set_title(f"Electron density at timestep {frame}")
ax.set_xlabel("x  (radial index)")
ax.set_ylabel("z  (along field line index)")
plt.show()
```
