## List Files

Checking available files. 'Fast' files are expected to be probes.

```python
all_files = list(data_dir.glob("*.npy"))
fast_files = list(data_dir.glob("*Fast*.npy"))
other_files = [f for f in all_files if f not in fast_files]

print(f"Total .npy files: {len(all_files)}")
print(f"Fast (Probe) files: {len(fast_files)}")
print(f"Other (Simulation) files: {len(other_files)}")

print("\nFast Files:")
for f in sorted(fast_files):
    print(f.name)

print("\nOther Files:")
for f in sorted(other_files):
    print(f.name)
```

## Check Data Shapes and Empty Files

```python
print("Checking Fast files for content and shapes...")
for f in sorted(fast_files):
    try:
        data = np.load(f)
        if data.size == 0:
            print(f"[EMPTY] {f.name}")
        else:
            print(f"[OK]    {f.name}: {data.shape}")
    except Exception as e:
        print(f"[ERROR] {f.name}: {e}")
```

```python
print("\nChecking Simulation files for content and shapes...")
for f in sorted(other_files):
    try:
        data = np.load(f)
        print(f"[OK]    {f.name}: {data.shape}")
    except Exception as e:
        print(f"[ERROR] {f.name}: {e}")
```
